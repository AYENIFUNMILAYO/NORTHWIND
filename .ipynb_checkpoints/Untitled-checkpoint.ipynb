# In the cell bellow, we want to know the quantity of products that is sold in each country by an employee 

In the table bellow: i created a new cloumn called peresent date, using 1-1-2021(note that northwind database was not released on 2021)in other to derived the age and year of service of the employee.

```python
## THE EMPLOYEE DATA THAT SHOWS THE AGE AND YEAR OF SERVICE WITHIN THE COMPLANY 
cur.execute("""SELECT concat(employees.last_name, ' ', employees.first_name) as Full_name, 
date_part('year', age(employees.present_date::date, employees.birth_date::date)) AS Ages, 
date_part('year', age(employees.present_date::date, employees.hire_date::date)) AS service_year,
sum(order_details.quantity) as sum_quantity, customers.country
FROM employees
JOIN orders ON orders.employee_id = employees.employee_id 
JOIN customers ON customers.customer_id = orders.customer_id
JOIN order_details ON order_details.order_id = orders.order_id
WHERE customers.country in ('USA','Portugal', 'Germany', 'Italy' )  
GROUP by customers.country,Full_name, Ages,service_year
ORDER BY sum_quantity """)




EMPLOYEES_SALES= pd.DataFrame(cur.fetchall(), columns=["Full_name","Ages", "service_year", "sum_quantity", "country"  ])

```

```python
## THE LEAST COUNTRY WHERE PRODUCT ARE ORDERED
cur.execute("""SELECT orders.order_id, customers.customer_id, customers.country, date_part('days', age(orders.shipped_date::date, orders.order_date::date)) AS waiting_period, 
date_part('days', age(orders.required_date::date, orders.shipped_date::date)) AS required_time,
order_details.quantity
from orders
JOIN customers ON customers.customer_id = orders.customer_id
JOIN order_details ON order_details.order_id = orders.order_id""")




sales_details= pd.DataFrame(cur.fetchall(), columns=["order_id","customer_id", "country","waiting_period","required_time", "quantity"])
```

## THE PERFORMANCE OF PRODUCTS IN EACH COUNTRY

# NOTE:

```python
cur.execute("""SELECT customers.country, products.product_name, sum(order_details.quantity) as Quantities
FROM order_details
JOIN orders 
ON orders.order_id = order_details.order_id
JOIN customers 
ON customers.customer_id = orders.customer_id
JOIN products 
ON products.product_id = order_details.product_id
group by customers.country, products.product_name
order by Quantities""")



country_productName = pd.DataFrame(cur.fetchall(), columns=["country", "product_name", "Quantities"])
```

```python
country_productName.head()
```

```python
def get_country_code(country_name):
    try:
        country = pycountry.countries.search_fuzzy(country_name)[0]
        return country.alpha_3
    except:
        return None
```

```python
# use the apply method to create a new column with the ISO code for each country
country_productName["country_code"] = country_productName["country"].apply(get_country_code)
```

```python
country_productName.head(9)
```

```python
country_productName['country'].value_counts()
```

```python
# load the world map data
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
```

```python
# merge the world map data with the country data based on the country code
merged = world.merge(country_productName, left_on='iso_a3', right_on='country_code', how='left')
```

```python
merged.head(5)
```

```python
merged['country'].value_counts()
```

```python
print(merged.columns)
```

```python
merged = merged.dropna(subset=['Quantities'])
```

```python
merged.head()
```

```python
merged.info()
```

```python
# Plot population estimates with an accurate legend
from mpl_toolkits.axes_grid1 import make_axes_locatable

fig, ax = plt.subplots(1, 1)

divider = make_axes_locatable(ax)

cax = divider.append_axes("right", size="5%", pad=0.1)

merged.plot(column='Quantities', ax=ax,cax=cax, legend=True)
ax.set_title('Product Purchase by Country')
plt.show()
```

## what week day does customer order more?
## what month does the customer order more?

```python
cur.execute("""SELECT to_char(order_date, 'Day') AS Day_Of_Week, 
to_char(order_date, 'Month') AS Months,customer_id
FROM orders
GROUP by customer_id, to_char(order_date, 'Day'), to_char(order_date, 'Month')""")



Days_Months = pd.DataFrame(cur.fetchall(), columns=[ "Day_Of_Week", "Months", "customer_id"])
```

```python
Days_Months.head()
```

```python
Days_Months.info()
```

```python
Days_Months['Day_Of_Week'] = Days_Months['Day_Of_Week'].astype('category')
Days_Months['Months'] = Days_Months['Months'].astype('category')
```

```python
Days_Months.info()
```

```python
Days_Months['Day_Of_Week'].value_counts()
```

```python
Days_Months['Months'].value_counts()
```

```python
##

fig,axes = plt.subplots(1,2, figsize=(16,6))
plt.tight_layout()
xlabels = Days_Months.Day_Of_Week
axes[0].set_title("Days Of Order ")
axes[0].set_xticklabels(xlabels, rotation=45, ha = 'right')
sns.countplot(x = 'Day_Of_Week', data = Days_Months, ax=axes[0])
axes[0].set_ylabel('Counts')
axes[0].set_ylabel('counts')

xlabels = Days_Months.Months
axes[1].set_title("Months Of The Year")
axes[1].set_xticklabels(xlabels, rotation=45, ha = 'right')
sns.countplot(x = 'Months', data = Days_Months, ax=axes[1])
axes[1].set_ylabel('Counts')
axes[1].set_ylabel('counts');
```

```python
cur.execute("""SELECT customers.customer_id, customers.country, round(avg(order_details.unit_price))as price_per_quantity, products.product_name, sum(order_details.quantity) as Quantities, sum(order_details.quantity*order_details.unit_price) as purchase_price
FROM order_details
JOIN orders 
ON orders.order_id = order_details.order_id
JOIN customers 
ON customers.customer_id = orders.customer_id
JOIN products 
ON products.product_id = order_details.product_id
group by customers.country, products.product_name,customers.customer_id

order by Quantities """)

df1= pd.DataFrame(cur.fetchall(), columns=["customer_id", "country","price_per_quantity", "product_name", "Quantities", "purchase_price"])
```

```python
df1.head()
```

```python
# merging df1 and days_months dataframe togethher
CA = pd.merge(Days_Months,df1, on='customer_id', how='outer')
```
